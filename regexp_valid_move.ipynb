## Pattern from google/gemini:

```python
pattern = r"[NBKRQ]?[a-h]?[1-8]?x?[a-h][1-8](=[NBKRQ])?[+#]?$|^[O0]-[O0](-[O0])?[+#]?"
```

```python
import re
```

```python
tests = ['q4','a5','fxg7','gxh7+','Nf6','Nxf6'\
         'Qxg7+']
```

```python
p_move = re.compile(pattern)
```

```python
for test in tests:
    q = p_move.search(test)
    if q is None:
        print("not a match: %s" % test)
```

```python
valid_move_pat = r"[NBKRQ]?[a-h]?[1-8]?x?[a-h][1-8](=[NBKRQ])?[+#]?|[O0]-[O0](-[O0])?[+#]?"
```

```python
best_move_pat = valid_move_pat + r" was best\."
best_move_pat
```

```python
best_move = re.compile(best_move_pat)
```

```python
test2 = "Bxg7 was best."
```

```python
best_move.search(test2)
```

```python
p_move.search(test2)
```

```python
p2 = re.compile(r"[NBKRQ]?[a-h]?[1-8]?x?[a-h][1-8](=[NBKRQ])?[+#]? was best\.")
```

```python
p2.search(test2)
```

## Regexp `subs` for valid latex output

```python
linein = "Re1# is the threat."
```

```python
re.sub(r"([NBKRQ]?[a-h]?[1-8]?x?[a-h][1-8](=[NBKRQ])?[+#]?|[O0]-[O0](-[O0])?[+#]?)", r"\\verb|\1|", linein)
```

```python
line2 = linein.replace('#','\\#')
line2
```
